```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset into a data frame using Python
df=pd.read_csv('Video games.csv')
df
```

```python
# Clean the data as needed
#print(df.head().isnull())
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
# df["Year"].fillna(df["Year"].mean(),inplace=True)
df["Year"] = df["Year"].fillna(df["Year"].mean())
df.isnull().sum()
```

```python
#Supprimer la colonne 'Name' du DataFrame
df=df.drop(columns=['Name'], axis=1)
```

```python
df.describe()
```

```python
# Compter le nombre de lignes dupliquÃ©es dans le DataFrame
print(f"Nombre de lignes dupliquÃ©es : {df.duplicated().sum()}")

# Comptage des lignes dupliquÃ©es (True pour dupliquÃ©es, False pour uniques)
df.duplicated().value_counts()
```

```python
import plotly.express as px
# 3.Create an area chart to show the trend of global sales over the years.
fig = px.area(df, x="Year", y="Global_Sales")
fig.show()
```

```python
import plotly.express as px
# 4.Create a histogram to show the distribution of global sales.
fig = px.histogram(df, x='Global_Sales')
fig.show()
```

```python
# 5.Create a scatter plot to show the relationship between North American sales and European sales.
plt.scatter(df['NA_Sales'], df['EU_Sales'])
plt.title('Relationship Between North American Sales and European Sales')
plt.xlabel('North American Sales')
plt.ylabel('European Sales')
plt.show()
```

```python
# 6.Create a scatter plot to show the relationship between Japanese sales and Other sales.
plt.scatter(df['JP_Sales'], df['Other_Sales'])
plt.title('Relationship between Japanese sales and Other sales')
plt.xlabel('Japanese sales')
plt.ylabel('Other sales')
plt.show()
```

```python
# 7.Create a stacked area chart to show the distribution of sales by platform over the years
df1 = df.groupby(['Year', 'Platform'])['Global_Sales'].sum().unstack()
df1.plot(kind='area', stacked=True, figsize=(10, 6))

plt.title('Distribution of Sales by Platform Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Select the top 10 publishers by Global sales
df2=df.groupby('Publisher')['Global_Sales'].sum().reset_index()
df2
df2 = df2.sort_values(by=['Global_Sales'], ascending=False)
top_10_publishers = df2.head(10)
top_10_publishers

sns.barplot(data=top_10_publishers, x='Global_Sales', y='Publisher', hue='Publisher', palette='Blues_d', legend=False)
plt.title('Top 10 Publishers by Global Sales')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Publisher')
plt.show()
```

```python
# 9.Create a pie chart to show the distribution of genres by global sales.

df3 = df.groupby('Genre')['Global_Sales'].sum()
df3

# Create a pie chart
plt.pie(df3, labels=df3.index, autopct='%1.1f%%')
plt.title('Distribution of Genres by Global Sales')
plt.show()
```
